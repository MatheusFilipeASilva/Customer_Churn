```python
from catboost import CatBoostClassifier
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, recall_score

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import joblib
```

#### Nesse notebook, geraremos um modelo simplificado, sem GridSearch, sem StratifiedKFold, apenas pra gerar nossa primeira pravisão e submissão, e avaliar o progresso do nosso modelo ao longo do percurso.

```python
df = joblib.load('df_att.pkl')
index = df['id']
```

```python
def extract_xy(df):
    target = 'Churn'
    X = df.drop([target, 'id'], axis=1)
    y = df[target]
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)

    return X_train, X_val, y_train, y_val

```

```python
X_train, X_val, y_train, y_val = extract_xy(df)

cat_features = X_train.select_dtypes(exclude='number').columns.tolist()
num_features = X_train.select_dtypes(include='number').columns.tolist()
```

```python
model = CatBoostClassifier(iterations = 200, learning_rate=0.6, depth=6, l2_leaf_reg=1.2, border_count=150, eval_metric='AUC')
```

```python
model.fit(X_train, y_train, cat_features=cat_features, eval_set=(X_val, y_val), verbose=100)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]

y_pred
```

```python
test = joblib.load('test_att.pkl')
index = test['id']
```

```python
test = test.drop('id', axis=1)
```

```python
test_predict = model.predict_proba(test)[:,1]
test_predict
```

```python
df_test = pd.DataFrame({'id': index, 'Churn': test_predict})

df_test.head(3)
```

```python
df_test.to_csv('1.0_submission.csv', index=False)
```

```python
predict = model.predict(X_val)
```

```python
ConfusionMatrixDisplay.from_predictions(y_val, predict, normalize='true')
```

```python
recall = recall_score(y_val, predict)
```
